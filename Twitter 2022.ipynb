```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("IPL_2022_tweets.csv")
df

```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])
df
```

```python
print(df.isnull())
```

```python
tweets_per_day = df.groupby(df['date'].dt.date).size()
```

```python
plt.figure(figsize=(10, 6))
tweets_per_day.plot(marker='o')
plt.title('Number of Tweets per Day')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
df['hashtags'] = df['hashtags'].astype(str)
```

```python
import re
# Preprocess hashtags: remove punctuation and split by whitespace
all_hashtags = ' '.join(df['hashtags'])
all_hashtags = re.sub(r'[^\w\s]', '', all_hashtags)  # Remove punctuation
all_hashtags = all_hashtags.split()  # Split by whitespace

# Count the frequency of each hashtag
word_counts = {}
for hashtag in all_hashtags:
    word_counts[hashtag] = word_counts.get(hashtag, 0) + 1

# Create the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import re
df['text'] = df['text'].astype(str)
df['hashtags'] = df['hashtags'].astype(str)

stop_words = ['a', 'an', 'the', 'of', 'on', 'in', 'and', 'is', 'are', 'to', 'for', 'with', 'you', 'your', 'this', 'that', 'it', 'at', 'by', 'from', 'or','will']

combined_text = ' '.join(df['text']) + ' ' + ' '.join(df['hashtags'])

# Preprocess combined text: remove punctuation and convert to lowercase
combined_text = re.sub(r'[^\w\s]', '', combined_text)  # Remove punctuation
combined_text = combined_text.lower()  # Convert to lowercase

# Split the combined text into words
words = combined_text.split()

filtered_words = [word for word in words if word not in stop_words]

# Count the frequency of each word
word_counts = {}
for word in filtered_words:
    word_counts[word] = word_counts.get(word, 0) + 1

# Convert word counts dictionary to DataFrame
word_counts_df = pd.DataFrame(word_counts.items(), columns=['Word', 'Frequency'])
word_counts_df = word_counts_df.sort_values(by='Frequency', ascending=False)
# Print the DataFrame
print(word_counts_df)
```

```python
top_20_words = word_counts_df.head(20)
top_20_words
```

```python
top_20_words = word_counts_df.head(20)

plt.figure(figsize=(10, 6))
plt.bar(top_20_words['Word'], top_20_words['Frequency'])
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Top 20 Most Used Words')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
